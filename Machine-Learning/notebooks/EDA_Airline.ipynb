```python
df.describe()
df.info()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0]) #will give us only the time by removing the things after the space and giving our index 0 value which is the time
df.head(2)
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops' :3, '4 stops' :4, np.nan:1})
df.head(2)
```

```python
df['Duration_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head(10)
```

```python
df['Duration_min'] = df['Duration_min'].fillna(0)
df.head(3)
```

```python
df['Duration_min'] = df['Duration_min'].astype(int)
df['Duration_hr'] = df['Duration_hr'].astype(int)
df.info()
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
encoded
```

```python
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoded_df
```

```python
df.drop(['Airline','Source','Destination'],axis=1,inplace=True)
pd.concat([encoded_df,df],axis=1)
```
