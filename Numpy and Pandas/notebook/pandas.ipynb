```python
import numpy as np
import pandas as pd

np.random.seed(123)
n_samples = 1000
girl_ratio = 0.9
girl_number = int(n_samples * girl_ratio)
boy_number = n_samples - girl_number
```

```python
#This will basically create 900 and 100 datapoints with random values, and the 0s and 1s would also be assigned to them
girl = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0, scale=1, size=girl_number),
    'feature_2' : np.random.normal(loc=0, scale=1, size=girl_number),
    'target' : [0] * girl_number
})

boy = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2, scale=1, size=boy_number),
    'feature_2' : np.random.normal(loc=2, scale=1, size=boy_number),
    'target' : [1] * boy_number
})
```

```python
df = pd.concat([girl,boy]).reset_index(drop=True)
df.tail()
```

```python
df.tail() #so we have created two features 1 and 2,both for boys and girls, that has completely random variables,now wach random variable is also assigned with either 0 for girl feature and 1 for boy feature 
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.9],random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X, columns=['feature_1','feature_2'])
df2 = pd.DataFrame(y, columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df['target'].value_counts()
```

```python

import matplotlib.pyplot as plt
plt.scatter(final_df['feature_1'],final_df['feature_2'],c=final_df['target'])
plt.show
```

```python
#SMOTE - interpolation
from imblearn.over_sampling import SMOTE
over_sampling = SMOTE()
X,y = over_sampling.fit_resample(final_df[['feature_1','feature_2']],final_df['target'])
```

```python
data = np.array([-250,-100,10,20,30,30,30,40,35,40,40,40,100,150])
import seaborn as sns
sns.boxplot(data)
```

```python
min,Q1,median,Q3,max = np.quantile(data,[0,0.25,0.5,0.75,1.0])
min,Q1,median,Q3,max
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1 - 1.5*(IQR)
upper_fence = Q3 + 1.5*(IQR)
```

```python
clean_data = data[(data >= lower_fence) & (data <= upper_fence)]
outliers = data[(data < lower_fence) | (data > upper_fence)]

print("Lower fence:", lower_fence)
print("Upper fence:", upper_fence)
print("Outliers:", outliers)
print("Clean data:", clean_data)
```
